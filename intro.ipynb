```python
import pandas as pd
import numpy as np
```

## filter session data (tokyo)

```python


# Load the CSV file from data/5minjan_aug/5min_6aug_23aug.csv
csv_path = 'data/5minjan_aug/5min_6aug_23aug.csv'
df = pd.read_csv(csv_path)
df['time'] = pd.to_datetime(df['time'], unit='s', utc=True)

# 3. Extract hour from datetime (UTC-based)
df['hour'] = df['time'].dt.hour

# 4. Define Tokyo session (00:00 to 09:00 UTC)
tokyo_df = df[(df['hour'] >= 0) & (df['hour'] < 9)].copy()

# 5. Drop helper column if not needed
tokyo_df.drop(columns=['hour'], inplace=True)

tokyo_df.head()
```

```python
# Assume tokyo_df is already created (filtered Tokyo session data)

# 1. Create a date column (based on UTC date)
tokyo_df['date'] = tokyo_df['time'].dt.date
tokyo_df['day_name'] = tokyo_df['time'].dt.day_name()

# 2. Group by each day to get High & Low
daily_stats = tokyo_df.groupby(['date', 'day_name']).agg(
    High=('high', 'max'),
    Low=('low', 'min')
).reset_index()

# 3. Calculate difference in raw price
daily_stats['Difference'] = daily_stats['High'] - daily_stats['Low']

# 4. Convert to points and pips
# For XAUUSD:
# 1 point = 0.01
# 1 pip   = 0.1
daily_stats['Difference_points'] = (daily_stats['Difference'] / 0.01).astype(int)
daily_stats['Difference_pips'] = (daily_stats['Difference'] / 0.1).astype(int)

# 5. Arrange columns as required
final_df = daily_stats[['date', 'day_name', 'High', 'Low', 
                        'Difference', 'Difference_points', 'Difference_pips']]

# Show result
final_df
```
